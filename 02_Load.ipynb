### Connect to local database

```python

rds_connection_string = "postgres:"+pw+"@localhost:5432/bigmac_database"
engine                = create_engine(f'postgresql://{rds_connection_string}')
```

### Check for tables

```python
engine.table_names()
```

### BIGMAC Use pandas to load csv converted DataFrame into database

```python
Combined_Big_Mac_Tables.to_sql(name='combined_big_mac_table', con=engine, if_exists='append', index=False)
```

### Confirm data has been added by querying the customer_name table
* NOTE: can also check using pgAdmin

```python
pd.read_sql_query('select * from combined_big_mac_table', con=engine).head()
```
